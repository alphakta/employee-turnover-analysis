```python
# Import libraries
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
```

```python
# Label encoding
clean_dataset.loc[:, "Education"] = LabelEncoder().fit_transform(clean_dataset["Education"])
clean_dataset.loc[:, "City"] = LabelEncoder().fit_transform(clean_dataset["City"])
clean_dataset.loc[:, "Gender"] = LabelEncoder().fit_transform(clean_dataset["Gender"])
clean_dataset.loc[:, "EverBenched"] = LabelEncoder().fit_transform(clean_dataset["EverBenched"])
```

```python
# Train-test split
X = clean_dataset.drop(columns = "LeaveOrNot")
y = clean_dataset["LeaveOrNot"]
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state = 1000, test_size = 0.20)
```

```python
# Oversampling
smote = SMOTE(sampling_strategy = "auto", random_state = 1000)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
# Build and train KNN model
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train_resampled, y_train_resampled)

# Confusion matrix of KNN model
conf_matrix = confusion_matrix(y_test, knn.predict(X_test))
sns.heatmap(conf_matrix, annot = True, fmt = "d", cmap = "Blues",
           xticklabels = ["Negative","Positive"],
           yticklabels = ["Negative", "Positive"])

# Evaluation of KNN model
print("KNN Model")

## Accuracy score
accuracy = accuracy_score(y_test, knn.predict(X_test))
print("Accuracy score:", accuracy)

## Classification score
classification_rep = classification_report(y_test, knn.predict(X_test))
print(classification_rep)
```

```python
# Build and train SVM model
svm = SVC(kernel = "linear")
svm.fit(X_train_resampled, y_train_resampled)

# Confusion matrix of SVM model
conf_matrix = confusion_matrix(y_test, svm.predict(X_test))
sns.heatmap(conf_matrix, annot = True, fmt = "d", cmap = "Blues",
           xticklabels = ["Negative","Positive"],
           yticklabels = ["Negative", "Positive"])

# Evaluation of SVM model
print("SVM Model")

## Accuracy score
accuracy = accuracy_score(y_test, svm.predict(X_test))
print("Accuracy score:", accuracy)

## Classification score
classification_rep = classification_report(y_test, svm.predict(X_test))
print(classification_rep)
```

```python
# Build and train Decision Tree model
dec_tree = DecisionTreeClassifier()
dec_tree.fit(X_train_resampled, y_train_resampled)

# Confusion matrix of Decision Tree model
conf_matrix = confusion_matrix(y_test, dec_tree.predict(X_test))
sns.heatmap(conf_matrix, annot = True, fmt = "d", cmap = "Blues",
           xticklabels = ["Negative","Positive"],
           yticklabels = ["Negative", "Positive"])

# Evaluation of Decision Tree model
print("Decision Tree Model")

## Accuracy score
accuracy = accuracy_score(y_test, dec_tree.predict(X_test))
print("Accuracy score:", accuracy)

## Classification score
classification_rep = classification_report(y_test, dec_tree.predict(X_test))
print(classification_rep)
```
